```python
import heapq

class Solution(object):
    def maxProduct(self, nums):
        a, b = heapq.nlargest(2, nums)
        return (a - 1) * (b - 1)
```

```python
nums = [3, 4, 5, 2]
obj = Solution()
result = obj.maxProduct(nums)
print(result)
```

```python
class Solution(object):
    def numTeams(self, rating):
        res = 0

        for j in range(len(rating)):
            ls = lg = rs = rg = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    ls += 1
                else:
                    lg += 1

            for k in range(j + 1, len(rating)):
                if rating[k] > rating[j]:
                    rg += 1
                else:
                    rs += 1

            res += ls * rg + lg * rs

        return res
```

```python
rating = [2, 5, 3, 4, 1]
obj = Solution()
result = obj.numTeams(rating)
print(result)
```

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
        busy = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime and endTime[i] >= queryTime:
                busy += 1
        return busy
```

```python
startTime = [1, 2, 3]
endTime = [3, 2, 7]
queryTime = 4

obj = Solution()
result = obj.busyStudent(startTime, endTime, queryTime)
print(result)
```

```python
class Solution(object):
    def numberOfSteps(self, num):
        if num == 0:
            return 0
        if num % 2 == 0:
            return 1 + self.numberOfSteps(num // 2)
        return 1 + self.numberOfSteps(num - 1)
```

```python
num = 14
obj = Solution()
result = obj.numberOfSteps(num)
print(result)
```

```python
class Solution(object):
    def countBits(self, n):
        ans = [0] * (n + 1)

        for i in range(1, n + 1):
            x = i
            while x:
                ans[i] += 1
                x &= x - 1

        return ans
```

```python
n = 5
obj = Solution()
result = obj.countBits(n)
print(result)
```
