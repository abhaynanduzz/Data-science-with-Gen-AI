```python
class Solution(object):
    def defangIPaddr(self, address):
        return "[.]".join(address.split("."))
```

```python
obj = Solution()
result = obj.defangIPaddr("1.1.1.1") 
print(result)
```

```python
class Solution(object):
    def findNumbers(self, nums):
        return sum(1 for n in nums if len(str(abs(n))) % 2 == 0)
```

```python
obj = Solution()           
result = obj.findNumbers([12,345,2,6,7896])
print(result)
```

```python
from collections import Counter

class Solution(object):
    def numIdenticalPairs(self, nums):
        freq = Counter(nums)
        total = 0
        for v in freq.values():
            total += v * (v - 1) // 2
        return total
```

```python
obj = Solution()
result = obj.numIdenticalPairs([4,2,3,4,4,3])
print(result)
```

```python
class Solution(object):
    def smallerNumbersThanCurrent(self, nums):
        freq = [0] * 101

        for n in nums:
            freq[n] += 1

        for i in range(1, 101):
            freq[i] += freq[i - 1]

        return [freq[n - 1] if n > 0 else 0 for n in nums]
```

```python
obj = Solution()
result = obj.smallerNumbersThanCurrent([8,1,2,2,3])
print(result)
```

```python
class Solution(object):
    def subtractProductAndSum(self, n):
        if n < 10:
            return n - n
        digit = n % 10
        rest = self.subtractProductAndSum(n // 10)
        return digit * (n // 10 if n // 10 else 1) - (digit + (n // 10))
```

```python
obj = Solution()
result = obj.subtractProductAndSum(191)
print(result)
```

```python
from functools import reduce

class Solution(object):
    def xorOperation(self, n, start):
        return reduce(lambda x, y: x ^ y,
                      [start + 2 * i for i in range(n)])
```

```python
obj = Solution()
print(obj.xorOperation(4, 0))
```
